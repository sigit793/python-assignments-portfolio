```python
#question 3
import numpy as np
import matplotlib.pyplot as plt

# Define normal density function
def normal_density(mean, variance, x):
    return (1 / np.sqrt(2 * np.pi * variance)) * np.exp(-((x - mean) ** 2) / (2 * variance))


x_values = np.linspace(-10, 10, 400)  
mean_01 = 0
variance_01 = 1
mean_02 = 2
variance_02 = 4


y_values_1 = normal_density(mean_1, variance_1, x_values)
y_values_2 = normal_density(mean_2, variance_2, x_values)

plt.plot(x_values, y_values_1, label=f'N({mean_1}, {variance_1})')
plt.plot(x_values, y_values_2, label=f'N({mean_2}, {variance_2})')
plt.legend()
plt.title('Normal Distribution Curves')
plt.xlabel('x')
plt.ylabel('f_X(x)')
plt.show()
```

```python
from scipy.integrate import quad

# Numerical integration function
def integration(mean, variance, a, b):
    result, _ = quad(lambda x: normal_density(mean, variance, x), a, b)
    return result

#example 
mean = 171
variance = 7.12  # in cm^2
a = 162
b = 190

probability = integration(mean, variance, a, b)
print(f"Probability of height between {a} and {b} cm: {probability:.4f}")
```

```python
#question 4a
# Uniform distribution expected value
def uniform_expectation(a, b):
    return (a + b) / 2

# Example: For interval [1, 3]
a = 1
b = 3
expected_value = uniform_expectation(a, b)
print(f"Expected value of uniform distribution on [{a}, {b}]: {expected_value}")
```

```python
# question 4b
def exponential_expectation(lam):
    return 1 / lam

lam = 1 / 50
expected_value = exponential_expectation(lam)
print(f"Expected time between pandemics: {expected_value} years")
```

```python
# question 4c
def normal_expectation_squared(mean, variance):
    return variance + mean**2

# average dosage
mean_height = 171
variance_height = 7.12
expected_X_squared = normal_expectation_squared(mean_height, variance_height)
average_dosage = 2.38 * expected_X_squared
print(f"Average dosage of drug: {average_dosage} units")
```
